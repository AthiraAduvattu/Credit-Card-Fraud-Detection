```python
#Oversampling (consider original data)
```

```python
X=data.drop('Class',axis=1)
y=data['Class']
```

```python
X.shape
```

```python
y.shape
```

```python
pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res,y_res=SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.2,random_state=42)
```

```python
#Logistic Regression after over sampling
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_pred1=log.predict(X_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
precision_score(y_test,y_pred1)
```

```python
recall_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```

```python
#Decision Tree Classifier after over sampling
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
y_pred2=dt.predict(X_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
precision_score(y_test,y_pred2)
```

```python
recall_score(y_test,y_pred2)
```

```python
f1_score(y_test,y_pred2)
```

```python
#Random Forest Classifier after over sampling
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_pred3=rf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred3)
```

```python
precision_score(y_test,y_pred3)
```

```python
recall_score(y_test,y_pred3)
```

```python
f1_score(y_test,y_pred3)
```

```python
final_data=pd.DataFrame({'Model':['LR','DT','RF'],'Acc_score':[accuracy_score(y_test,y_pred1),accuracy_score(y_test,y_pred2),accuracy_score(y_test,y_pred3)]})
```
